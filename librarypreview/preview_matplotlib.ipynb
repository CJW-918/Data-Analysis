```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 주피터노트북에서 그래프 출력 가능하게 함.
%matplotlib inline
```

```python
names = ['Bob', 'Jessica', 'Mary', 'John', 'Mel']
births = [123,456, 789, 112, 345]
custom = [1, 5, 25, 13, 23232]

data_set = list(zip(names, births, custom))
df = pd.DataFrame(data = data_set, columns = ['Names', 'births', 'custom'])
```

```python
x = df['Names']
y = df['births']

# 막대그래프 출력
plt.bar(x, y)
plt.xlabel('Names')
plt.ylabel('births')
plt.title('bar print')
plt.show()
```

```python
np.random.seed(19920613)

X = np.arange(0.0, 100.0, 5.0)
Y = (X * 2.0) + np.random.rand(20) * 30
```

```python
# 산점도 데이터 출력
plt.scatter(X, Y, c='b', alpha = 0.5, label='scatter point')
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc='upper left')
plt.title('scatter point')
plt.show()
```
